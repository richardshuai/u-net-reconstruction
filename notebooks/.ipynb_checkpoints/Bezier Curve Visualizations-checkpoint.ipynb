```python
import numpy as np
import matplotlib.pyplot as plt
import skimage.draw
```

```python
# Visualizing how parameters change Bezier curve drawings.
Y, X = 50, 50
T = 10

weights = np.linspace(0, 1, T)
y1 = np.random.randint(Y)
x1 = np.random.randint(X)

y2 = np.random.randint(Y)
x2 = np.random.randint(X)

y3 = np.random.randint(Y)
x3 = np.random.randint(X)
for weight in weights:
    image = np.zeros((Y, X))
    curve = skimage.draw.bezier_curve(y1, x1, y2, x2, y3, x3, weight)
    image[curve] = 1
    image[y1, x1] = 1.5
    image[y2, x2] = 1.5
    image[y3, x3] = 1.5
    plt.imshow(image)
    plt.title("({0}, {1}), ({2}, {3}), ({4}, {5}), weight: {6}".format(y1, x1, y2, x2, y3, x3, weight))
    plt.show()
    
```
