```python
import numpy as np
import tensorflow as tf

import scipy.io as sio
from os import path
from PIL import Image
from PIL import ImageOps
import matplotlib.pyplot as plt
```

```python
# Names of sample files: sample_names-xxxxx
sample_names = 'zebrafish'

record_dir = path.join('..', 'data', '2d_data', 'tf_records', 'zebrafish_and_beads', 'ground_truths')

# Target number of images per image, should be divisible by 3.
num_images_per = 99
```

```python
zebrafish_mat_file = path.join('../data/2d_data/real_data/fish_double.mat')
```

```python
zfs = np.array(sio.loadmat(zebrafish_mat_file)['fish'])
zfs = zfs.transpose(2, 0, 1)    # Transpose to achieve n, y, x order
```

```python
def _create_example(plane):
    """
    Creates and returns tf.Example from a given numpy array.
    """
    plane_feature = tf.train.Feature(float_list=tf.train.FloatList(value=plane.ravel()))
    feature = {
        'plane': plane_feature
    }
    return tf.train.Example(features=tf.train.Features(feature=feature))
```

```python
for i in range(len(zfs)):
    zf_image = Image.fromarray(zfs[i])
    zf_image_flipped = ImageOps.flip(zf_image)
    zf_image_mirrored = ImageOps.mirror(zf_image)
    
    num_rotations = num_images_per // 3
    degree_increment = 360 / num_rotations
    for j in range(num_rotations):
        num_degrees = degree_increment * j
        
        zf_1 = np.asarray(zf_image.rotate(num_degrees))
        filename_1 = sample_names + '-%.5d' % (i * num_images_per + 3*j)
        recordfile_1 = path.join(record_dir, filename_1)
        with tf.io.TFRecordWriter(recordfile_1) as writer:
            tf_example = _create_example(zf_1)
            writer.write(tf_example.SerializeToString())
        
        zf_2 = np.asarray(zf_image_flipped.rotate(num_degrees))
        filename_2 = sample_names + '-%.5d' % (i * num_images_per + 3*j + 1)
        recordfile_2 = path.join(record_dir, filename_2)
        with tf.io.TFRecordWriter(recordfile_2) as writer:
            tf_example = _create_example(zf_2)
            writer.write(tf_example.SerializeToString())
        
        zf_3 = np.asarray(zf_image_mirrored.rotate(num_degrees))
        filename_3 = sample_names + '-%.5d' % (i * num_images_per + 3*j + 2)
        recordfile_3 = path.join(record_dir, filename_3)
        with tf.io.TFRecordWriter(recordfile_3) as writer:
            tf_example = _create_example(zf_3)
            writer.write(tf_example.SerializeToString())
```
